write a python code for colab that plot a graph with linear function from -10 until 10, with offset 2 and incline 1

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the range of x from -10 to 10
x = np.linspace(-10, 10, 400)

# Define the linear function with incline 1 and offset 2
y = x + 2

# Plot the graph
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='y = x + 2')
plt.title('Linear Function: y = x + 2')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

add a UI of slider to the offset and incline so that I can manually change them

```python
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact

# Function to plot the graph based on given incline and offset
def plot_graph(incline=1, offset=2):
    x = np.linspace(-10, 10, 400)
    y = incline * x + offset

    plt.figure(figsize=(8, 6))
    plt.plot(x, y, label=f'y = {incline}x + {offset}')
    plt.title(f'Linear Function: y = {incline}x + {offset}')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid(True)
    plt.legend()
    plt.show()

# Create sliders for incline and offset
interact(plot_graph, incline=(-5.0, 5.0), offset=(-10.0, 10.0))
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the range of x and the linear function
x_range = np.linspace(-10, 10, 400)
incline = 1
offset = 2
y_line = incline * x_range + offset

# Generate random points
np.random.seed(0)  # For reproducibility
x_points = np.random.uniform(-10, 10, 100)
y_random = np.random.uniform(-10, 14, 100)

# Separate points above and below the line
above_line = y_random > incline * x_points + offset
below_line = ~above_line

# Plot the graph
plt.figure(figsize=(8, 6))
plt.plot(x_range, y_line, label='y = x + 2', color='black')
plt.scatter(x_points[above_line], y_random[above_line], color='red', label='Above line')
plt.scatter(x_points[below_line], y_random[below_line], color='blue', label='Below line')
plt.title('Random Points Relative to the Line y = x + 2')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

now write a function that insert this points to dataset so the fetures will be the x values and the predict will be the y

```python
import pandas as pd
import numpy as np

def create_dataset(incline=1, offset=2, n_points=50):
    # Generate 50 random points above and 50 below the line
    np.random.seed(0)  # for reproducible results
    x_above = np.random.uniform(-10, 10, n_points)
    y_above = incline * x_above + offset + np.random.uniform(1, 5, n_points)

    x_below = np.random.uniform(-10, 10, n_points)
    y_below = incline * x_below + offset - np.random.uniform(1, 5, n_points)

    # Combine the x and y values for both above and below points
    x_all = np.concatenate((x_above, x_below))
    y_all = np.concatenate((y_above, y_below))

    # Create a DataFrame with 'features' for x values and 'predict' for y values
    dataset = pd.DataFrame({
        'features': x_all,
        'predict': y_all
    })

    return dataset

# Create the dataset
dataset = create_dataset()
print(dataset)  # Display the first few rows of the dataset
```

write a python code for colab that plot 5 linear functions that create a sinos function, display on each linear function his offset and incline.
create 5 points on each line, 10 points below the line and 10 points above the line, mark in the graph the ranges of x and y for each linear function

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_sine_approximation():
    # Sine wave parameters
    periods = 1  # Number of periods to cover
    segments_per_period = 5  # Number of linear segments per period

    # Generate x values for the sine wave
    x_sine = np.linspace(-np.pi * periods, np.pi * periods, 1000)
    y_sine = np.sin(x_sine)

    # Determine the segments for the linear approximations
    x_segments = np.linspace(-np.pi * periods, np.pi * periods, segments_per_period * periods + 1)
    y_segments = np.sin(x_segments)

    # Plotting the sine wave
    plt.figure(figsize=(14, 8))
    plt.plot(x_sine, y_sine, label='Sine Wave', color='gray', linestyle='--')

    # Plotting the linear segments
    for i in range(len(x_segments) - 1):
        x_range = np.linspace(x_segments[i], x_segments[i+1], 100)
        incline = (y_segments[i+1] - y_segments[i]) / (x_segments[i+1] - x_segments[i])
        offset = y_segments[i] - incline * x_segments[i]
        y_range = incline * x_range + offset

        plt.plot(x_range, y_range, label=f'Line {i+1}: y = {incline:.2f}x + {offset:.2f}')

        # Points on, above, and below the line
        x_points = np.linspace(x_segments[i], x_segments[i+1], 5)
        y_points_on_line = incline * x_points + offset
        y_points_above = y_points_on_line + np.random.uniform(0.1, 0.3, 5)
        y_points_below = y_points_on_line - np.random.uniform(0.1, 0.3, 5)

        plt.scatter(x_points, y_points_on_line, color='black')
        plt.scatter(x_points, y_points_above, color='red')
        plt.scatter(x_points, y_points_below, color='blue')

        # Annotate the range, incline, and offset for each linear function
        plt.text((x_segments[i] + x_segments[i+1]) / 2, offset,
                 f'Range: x = {x_segments[i]:.2f} to {x_segments[i+1]:.2f}\nIncline: {incline:.2f}, Offset: {offset:.2f}',
                 fontsize=9, ha='center')

    # Graph details
    plt.title('Linear Approximations of a Sine Wave')
    plt.xlabel('x-axis')
    plt.ylabel('y-axis')
    plt.legend()
    plt.grid(True)
    plt.show()

plot_sine_approximation()
```
